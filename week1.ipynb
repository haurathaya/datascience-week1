```python
#import library yang digunakan

import pandas as pd
import numpy as np
```

```python
#import dataset

dataset = pd.read_excel('tugas_week1.xlsx')
```

```python
#menampilkan dataset

dataset
```

```python
#menampilkan data 6 teratas

dataset.head(6)
```

```python
#menampilkan data 6 terbawah

dataset.tail(6)
```

```python
#mengubah variable pada kolom gender. apabila F -> Perempuan, apabila M -> Laki-laki

dataset['gender'] = dataset.gender.replace(to_replace="M", value="Laki-laki")
dataset['gender'] = dataset.gender.replace(to_replace="F", value="Perempuan")

dataset
```

```python
#melihat outliers

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

mpl.rcParams['figure.dpi'] = 250
plt.figure(figsize=(60, 60))
f, axes = plt.subplots(1, 3)
sns.boxplot(y=dataset["tinggi_badan"], ax=axes[0])
sns.boxplot(y=dataset["berat_badan"], ax=axes[1])
sns.boxplot(y=dataset["ipk"], ax=axes[2])
plt.subplots_adjust(wspace=1)
```
